## Otp class

```python
class Otp:
  def __init__(self, txt, key = None):
    self.key = self.toUnicode(self.genKey(len(txt))) if key == None else self.toUnicode(key)
    self.txt = [x.lower() for x in txt]
    self.unicode_lst = self.toUnicode(self.txt)
    self.enc_txt = None
    self.dec_txt = None
    print(f"The unicode representation of the message is: {self.unicode_lst}")
    print(f"The unicode representation of the key is: {self.key}")

  def genKey(self, length):
    key = [chr(rd.randint(97, 122)) for i in range(length)]
    return key

  @staticmethod
  def toUnicode(char_list):
    unicode_list = [ord(x) for x in char_list]
    return unicode_list

  @staticmethod
  def toString(unicode_txt):
    char_list = "".join([chr(x) for x in unicode_txt])
    return char_list

  def encrypt(self):
    self.enc_txt = [self.unicode_lst[i] + self.key[i] for i in range(len(self.unicode_lst))]
    print(f"The unicode representation of the encrypted message is: {self.enc_txt}")
    char_list = self.toString(self.enc_txt)
    return char_list

  def decrypt(self):
    self.dec_txt = [self.enc_txt[i] - self.key[i] for i in range(len(self.enc_txt))]
    print(f"The unicode representation of the decrypted message is: {self.dec_txt}")
    char_list = self.toString(self.dec_txt)
    return char_list
```

## Code execution

```python
txt = input("Enter the message to be encrypted: ")
txt_key = input("Enter the key: ")
```

Execute mode with automatic key

```python
otp = Otp(txt)
print(f"The encrypted message is: {otp.encrypt()}")
print(f"The decrypted message is: {otp.decrypt()}")
```

Execute mode with manual key

```python
otp = Otp(txt, key = txt_key)
print(f"The encrypted message is: {otp.encrypt()}")
print(f"The decrypted message is: {otp.decrypt()}")
```
